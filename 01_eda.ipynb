```python
import pandas as pd

# Load raw reports
reports = pd.read_csv(r"K:\I-EHRs-Project\data\raw\indiana_reports.csv")
projections = pd.read_csv(r"K:\I-EHRs-Project\data\raw\indiana_projections.csv")

# Existing validation CSV
val_csv = pd.read_csv(r"K:\I-EHRs-Project\data\validation_questions.csv")

qa_list = []

# From indiana_reports.csv
for idx, row in reports.iterrows():
    # Combine findings + impression as answer
    report_text = str(row['findings']) + " " + str(row['impression'])
    
    qa_list.append({
        "question": "What abnormality is present in the right lung?",
        "gold_answer": report_text
    })
    qa_list.append({
        "question": "Is there any opacity seen?",
        "gold_answer": report_text
    })
    qa_list.append({
        "question": "Is the heart size normal?",
        "gold_answer": report_text
    })

# From indiana_projections.csv
for idx, row in projections.iterrows():
    proj_text = str(row['projection'])
    qa_list.append({
        "question": "What abnormality is present according to projection?",
        "gold_answer": proj_text
    })

# Append to existing validation file
qa_df = pd.DataFrame(qa_list)
full_val = pd.concat([val_csv, qa_df], ignore_index=True)
full_val.to_csv(r"K:\I-EHRs-Project\data\validation_questions.csv", index=False)

print(f"Validation file updated. Total QA pairs now: {len(full_val)}")
```
